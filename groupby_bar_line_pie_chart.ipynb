```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"zomato_dataset.csv")
print(df)
df.columns
df.info()
df.describe()
df.isnull().sum()
df.dropna(inplace=True)
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df["Total_Sales"] = df["Price"] * df["Quantity"]
print(df["Total_Sales"])
```

```python

df["Total_Sales"] = df["Price"] * df["Quantity"]
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Month"] = df["Order_Date"].dt.month
```

```python
print(df["Month"])
```

```python
# Region chart
df.groupby("City")["Total_Sales"].sum().plot(kind="barh")
plt.show()
```

```python
# Product chart
df.groupby("Cuisine")["Total_Sales"].sum().plot(kind="barh")
plt.show()
```

```python
# Restaurant
# Product chart
df.groupby("Restaurant")["Total_Sales"].sum().plot(kind="pie")
plt.show()
```

```python
# Month chart
df.groupby("Month")["Total_Sales"].sum().plot(kind="line")
plt.show()
```

```python
df["Total_Sales"] = df["Price"] * df["Quantity"]

Cuisine_sales = df.groupby("Cuisine")["Total_Sales"].sum()

plt.pie(Cuisine_sales, labels=Cuisine_sales.index, autopct='%1.2f%%')
plt.title("Cuisine-wise Sales Distribution")
plt.show()
```
